```python
def get_outliers(resid):
    sigma = np.std(resid)
    mu = np.mean(resid)
    outlier = [(((r-mu)/sigma)**2>3)*1*r/abs(r) for r in resid]
    return outlier
```

```python
def plot_altered(obj, observed=True, seasonal=True, trend=True, resid=True,
         weights=False, draw_ref = False):
    """
    Plot estimated components
    Parameters
    ----------
    observed : bool
        Include the observed series in the plot
    seasonal : bool
        Include the seasonal component in the plot
    trend : bool
        Include the trend component in the plot
    resid : bool
        Include the residual in the plot
    weights : bool
        Include the weights in the plot (if any)
    Returns
    -------
    matplotlib.figure.Figure
        The figure instance that containing the plot.
    """
    from statsmodels.graphics.utils import _import_mpl
    from pandas.plotting import register_matplotlib_converters
    plt = _import_mpl()
    register_matplotlib_converters()
    print('Using a reference value of {} for trend change detection'.format(obj.reference))
    series = [(obj.observed, 'Observed')] if observed else []
    assert len(series)
    series += [(obj.trend, 'trend')] if trend else []
    series += [(obj.seasonal, 'seasonal')] if seasonal else []
    series += [(obj.resid, 'residual')] if resid else []
    series += [(obj.weights, 'weights')] if weights else []
    if trend:
        #assert 'trend_flag' in obj
        trend_flag = obj.trend_flag
    if resid:
        #assert 'resid_flag' in obj
        resid_flag = obj.resid_flag
        
    if isinstance(obj.observed, (pd.DataFrame, pd.Series)):
        nobs = obj.observed.shape[0]
        xlim = obj.observed.index[0], obj.observed.index[nobs - 1]
    else:
        xlim = (0, obj.observed.shape[0] - 1)

    fig, axs = plt.subplots(len(series), 1)
    for i, (ax, (series, def_name)) in enumerate(zip(axs, series)):
        if def_name != 'residual':
            ax.plot(series)
            if def_name == 'trend':
                inds = trend_flag.loc[(trend_flag==1)].index
                for ind in inds:
                    ax.axvline(x=ind, color='black')
                inds = trend_flag.loc[(trend_flag==-1)].index
                for ind in inds:
                    ax.axvline(x=ind, color='red')
                
        else:
            inds = resid_flag.loc[(resid_flag==1)].index
            for ind in inds:
                ax.axvline(x=ind, color='black')
            inds = resid_flag.loc[(resid_flag==-1)].index
            for ind in inds:
                ax.axvline(x=ind, color='red')
            if draw_ref:
                ax.axhline(y=obj.reference, color='black')
                ax.axhline(y=-obj.reference, color='black')
            ax.plot(series)#, marker='o', linestyle='none')
            #ax.plot(xlim, (0, 0), color='#000000', zorder=-3)
        #name = getattr(series, 'name', def_name)
        name = getattr(obj, 'name')
        if def_name == 'Observed' and name is not None:
            use_name = copy(name)
            name = name.capitalize()
        else:
            use_name = copy(def_name)
        title = ax.set_title if i == 0 and observed else ax.set_ylabel
        title(use_name)
        ax.set_xlim(xlim)

    fig.tight_layout()
    #return fig
```
