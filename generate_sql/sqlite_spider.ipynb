```python
from sqlalchemy import *
from sqlalchemy.ext.automap import automap_base
import pandas as pd
import os
import json
import subprocess

path_to_data = '/home/eric/Triplet/data/spider/database/'
folders = os.listdir(path_to_data)
```

```python
final_dict = {}
for f in folders:
    try:
#         print('database: {}'.format(f))
        engine = sqlalchemy.create_engine('sqlite:///' + path_to_data + '{}/'.format(f) + '{}.sqlite'.format(f))
        Base = automap_base()
        Base.prepare(engine, reflect=True)
        inspector = inspect(engine)
        final_dict[f] = {}
        #get tables
        tables = []
        tables = Base.classes.keys()
        for t in tables:
            final_dict[f][t]={}
#             print('table: {}'.format(t))
            cols = inspector.get_columns('{}'.format(t))
            for c in cols:
                if (type(c['type'])==sqlalchemy.sql.sqltypes.TEXT) |  (
                    type(c['type'])==sqlalchemy.sql.sqltypes.VARCHAR):

                    h = c['name']
                    df = pd.read_sql('SELECT DISTINCT "{0}" FROM {1}'.format(h,t), engine)

                    final_dict[f][t][h] = list(df.iloc[:,0])



#         print('**'*50)
    except:
        print('failed at {}'.format(f))
#     break
```

```python
with open('/home/eric/Triplet/data/spider/uniques.json', 'w') as f:
    json.dump(final_dict, f)
```
